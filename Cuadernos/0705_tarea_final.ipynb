En este ejercicio vamos a seguir sin implementar nada nuevo; simplemente vamos a hacer nuestro código más compacto y eficiente, a evitar que haya código repetido creando funciones. Esto se conoce como _refactorizar_ el código.

Modifica el programa anterior, para que:
- Existan las siguientes funciones:
    - `preguntar_opcion()`: que imprime el menú y devuelve la opción elegida por el usuario
    - `pedir_datos()`: que pide los datos de perfil al usuario y los devuelve
    - `imprimir_datos()`: que imprime los datos de perfil del usuario
    - `crear_perfil()`: que llama a `pedir_datos()`, imprime el mensaje `Tu perfil ha sido creado` y llama a `imprimir_datos()`
    - `modificar_perfil()`: que llama a `pedir_datos()`, imprime el mensaje `Tu perfil ha sido modificado` y llama a `imprimir_datos()`
    - `publicar_mensaje()`: que pregunta e imprime el mensaje
- En el programa principal, existan llamadas solamente a `preguntar_opcion()`, `crear_perfil()`, `modificar_perfil()` y `publicar_mensaje()`, en el lugar correspondiente del bucle.
- Esté separado en 3 partes: datos (que contendrá el diccionario de provincias), funciones y programa principal, en ese orden.
- Esté bien documentado, es decir, haya comentarios separando las partes del programa y lo que hace cada función, al menos.
