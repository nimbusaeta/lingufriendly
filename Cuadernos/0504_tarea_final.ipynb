# Ejercicio final del tema 5 - Red social
Hasta ahora, el usuario tenía un poder de decisión muy limitado: solo podía hacer una cosa cada vez que ejecutaba el programa.

Pero ya hemos aprendido a hacer que el menú permanezca, y a que el programa no se acabe hasta que el usuario quiera, así que vamos a aplicarlo a nuestro proyecto.

Modifica el programa anterior para que esta vez:
- Haya una cuarta opción, `Salir`, en el menú de usuario. Cuando sea elegida, debe imprimirse un mensaje de despedida como `¡Hasta pronto!` y dejar de ejecutarse el programa.
- El usuario pueda crear perfiles, modificarlos y poner mensajes de estado todas las veces que quiera en una sola ejecución, hasta que elija la opción `Salir`. Cada vez que elija algo que no sea salir, se debe imprimir el menú de nuevo.
- En el caso de que el usuario elija una opción no presente en el menú, se imprima un recordatorio de las opciones válidas.
