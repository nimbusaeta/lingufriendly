# Ejercicio final del tema 3 - Red social
En el ejercicio del tema anterior, solo habíamos trabajado con datos del usuario de tipo string. Además no habíamos manipulado ninguno de ellos. Ahora, vamos a poner en práctica lo que hemos aprendido en este tema.

Modifica el programa anterior de forma que:
- Le pida al usuario el dato de su estatura en centímetros y después, en la parte de imprimir el perfil, la imprima expresada en metros.
- En vez de pedirle al usuario su edad, como antes, le pida su año de nacimiento. Después, en la parte de imprimir el perfil, que calcule su edad actual teniendo en cuenta que estamos en 2019, y la imprima.
- Corrija los apellidos del usuario. Es decir, que dé igual si el usuario los escribe todo en mayúsculas, todo en minúsculas o en mayúsculas y minúsculas con creatividad, que tu programa sea capaz de imprimirlo con las primeras en mayúsculas. Tendrás que preguntar separadamente por cada uno de los apellidos.

Debería quedar, por ejemplo, así:

    ¿Cómo te llamas?: Gustavo Adolfo Claudio
    ¿Cuál es tu primer apellido?: DOMÍNGUEZ
    ¿Y el segundo?: BASTIDA
    ¿Cuál es tu año de nacimiento?: 1836
    Y para acabar, ¿cuál es tu estatura? (en centímetros): 178
    
    Tu perfil ha sido creado.
    Nombre: Gustavo Adolfo Claudio Domínguez Bastida
    Edad: 183 años
    Estatura: 1.78 metros
    
    ¿Qué estás pensando?: ¿Qué es poesía?, dices, mientras clavas en mi pupila tu pupila azul
    
    Gustavo Adolfo Claudio dice: ¿Qué es poesía?, dices, mientras clavas en mi pupila tu pupila azul

Recuerda mantener la estructura del programa anterior. En programación se suele trabajar con código ya existente (lo que llaman _legacy code_) y simplemente ampliarlo o mejorarlo, así que debes acostumbrarte a toquetear código.
